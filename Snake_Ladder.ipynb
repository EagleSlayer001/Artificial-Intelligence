```python
import random
from collections import deque

def setup_board():
    board = [i for i in range(101)]
    ladders = {
        3: 22, 5: 8, 11: 26, 20: 29
    }
    snakes = {
        27: 1, 17: 4, 19: 7, 21: 9
    }
    for start, end in {**ladders, **snakes}.items():
        board[start] = end
    return board

def roll_dice():
    return random.randint(1, 6)

def play_game():
    board = setup_board()
    num_players = int(input("Enter the number of players: "))
    players = [1] * num_players
    player_names = [input(f"Enter name for player {i+1}: ") for i in range(num_players)]

    turns_per_player = 3
    player_turns = [0] * num_players
    winner = None
    turn = 0

    while all(turns < turns_per_player for turns in player_turns):
        current_player = turn % num_players
        if player_turns[current_player] < turns_per_player:
            input(f"{player_names[current_player]}'s turn! Roll the dice...")

            dice = roll_dice()
            print(f"{player_names[current_player]} rolled a {dice}!")

            next_square = players[current_player] + dice
            if next_square > 100:
                print(f"{player_names[current_player]} stays at {players[current_player]} (cannot move beyond 100).")
            else:
                final_square = board[next_square]
                print(f"{player_names[current_player]} moves from {players[current_player]} to {final_square}.")
                players[current_player] = final_square

                if players[current_player] == 100:
                    winner = player_names[current_player]
                    print(f"Congratulations! {winner} has won the game!")
                    return

            player_turns[current_player] += 1
        turn += 1

    max_position = max(players)
    winners = [player_names[i] for i, pos in enumerate(players) if pos == max_position]
    if len(winners) > 1:
        print(f"The game ended in a tie! Winners: {', '.join(winners)}")
    else:
        print(f"{winners[0]} is the winner based on the furthest position!")

play_game()
46
3
```
