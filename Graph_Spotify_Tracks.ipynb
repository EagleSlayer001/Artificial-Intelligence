```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    'track_name': ['Comedy', 'Ghost - Acoustic', 'To Begin Again', 'Can\'t Help Falling In Love', 'Hold On'],
    'popularity': [73, 55, 57, 71, 82]
}

df = pd.DataFrame(data)

# Plotting the graph
plt.figure(figsize=(10, 6))
plt.bar(df['track_name'], df['popularity'], color='skyblue')
plt.xlabel('Track Name')
plt.ylabel('Popularity')
plt.title('Popularity of Spotify Tracks')
plt.xticks(rotation=45)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    'track_name': ['Comedy', 'Ghost - Acoustic', 'To Begin Again', 'Can\'t Help Falling In Love', 'Hold On'],
    'popularity': [73, 55, 57, 71, 82],
    'release_year': [2019, 2020, 2021, 2018, 2022]
}

df = pd.DataFrame(data)

df = df.sort_values('release_year')

plt.figure(figsize=(10, 6))
plt.plot(df['release_year'], df['popularity'], marker='o', linestyle='-', color='red')
plt.xlabel('Year')
plt.ylabel('Popularity')
plt.title('Popularity of Spotify Tracks Over Time')
plt.xticks(df['release_year'])
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    'track_name': ['Comedy', 'Ghost - Acoustic', 'To Begin Again', 'Can\'t Help Falling In Love', 'Hold On'],
    'popularity': [73, 55, 57, 71, 82],
    'release_year': [2019, 2020, 2021, 2018, 2022]
}

df = pd.DataFrame(data)

plt.figure(figsize=(10, 6))
plt.scatter(df['release_year'], df['popularity'], color='brown')
plt.xlabel('Year')
plt.ylabel('Popularity')
plt.title('Scatter Plot of Song Popularity Over Years')
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    'track_name': ['Comedy', 'Ghost - Acoustic', 'To Begin Again', 'Can\'t Help Falling In Love', 'Hold On'],
    'popularity': [73, 55, 57, 71, 82]
}

df = pd.DataFrame(data)

plt.figure(figsize=(8, 8))
plt.pie(df['popularity'], labels=df['track_name'], autopct='%1.1f%%', startangle=90, colors=['#ff9999','#66b3ff','#99ff99','#ffcc99','#c2c2f0'])
plt.title('Popularity Distribution of Spotify Tracks')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data including the track names and popularity
data = {
    'track_name': ['Comedy', 'Ghost - Acoustic', 'To Begin Again', 'Can\'t Help Falling In Love', 'Hold On'],
    'popularity': [73, 55, 57, 71, 82]
}

df = pd.DataFrame(data)

# Plotting the histogram
plt.figure(figsize=(8, 6))
plt.hist(df['popularity'], bins=5, color='purple', edgecolor='black')
plt.xlabel('Popularity')
plt.ylabel('Frequency')
plt.title('Histogram of Spotify Track Popularity')
plt.grid(True)
plt.show()
```
